```python
import pandas as pd
import pydicom
import pylab
from PIL import Image
import skimage
from skimage.color import rgb2gray
from skimage.transform import resize
from skimage import filters
import numpy as np
import matplotlib.pyplot as plt
import os
import re
from tqdm.auto import tqdm, trange
```

```python
data_dir='/content/drive/MyDrive/Разметка изображений ЭХОКГ/DICOM_IMAGE/'
data=dict()
for file in tqdm(os.listdir(data_dir)):
    #print(file)
    ds= pydicom.dcmread(data_dir+file)
    img=ds.pixel_array
    i=img.shape
    data.setdefault('namber_img',[]).append(i[0])
    data.setdefault('rows',[]).append(i[1])
    data.setdefault('columns',[]).append(i[2])
    data.setdefault('name_file',[]).append(file)
    data.setdefault('label',[]).append(file.split('_')[1])
data1=pd.DataFrame(data)
data1
```
